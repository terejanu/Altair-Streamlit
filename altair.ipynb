# Altair Tutorial
[https://altair-viz.github.io/getting_started/starting.html#](https://altair-viz.github.io/getting_started/starting.html#)

```python
import altair as alt
import pandas as pd
import numpy as np
from vega_datasets import data
```

# Simple Bar Chart
[https://altair-viz.github.io/gallery/simple_bar_chart.html](https://altair-viz.github.io/gallery/simple_bar_chart.html)

```python
source_bc = pd.DataFrame({
    'a': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I'],
    'b': [28, 55, 43, 91, 81, 53, 19, 87, 52]
})

source_bc
```

```python
alt.Chart(source_bc).mark_bar().encode(
    x='a',
    y='b'
)

# Marks
# https://altair-viz.github.io/user_guide/marks.html
```

# Simple Scatter Plot
[https://altair-viz.github.io/gallery/scatter_tooltips.html](https://altair-viz.github.io/gallery/scatter_tooltips.html)

```python
source_sp = data.cars()

source_sp
```

```python
alt.Chart(source_sp).mark_circle(size=60).encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color='Origin',
    tooltip=['Name', 'Origin', 'Horsepower', 'Miles_per_Gallon']
).interactive()

# Encoding Channels
# https://altair-viz.github.io/user_guide/encoding.html#encoding-channels
```

# Simple Line Chart
[https://altair-viz.github.io/gallery/simple_line_chart.html](https://altair-viz.github.io/gallery/simple_line_chart.html)

```python
x = np.arange(100)
source_lc = pd.DataFrame({
  'x': x,
  'f(x)': np.sin(x / 5)
})

source_lc
```

```python
alt.Chart(source_lc).mark_line().encode(
    x='x',
    y='f(x)'
)
```

# Compound Charts - example electricity
[https://altair-viz.github.io/user_guide/compound_charts.html](https://altair-viz.github.io/user_guide/compound_charts.html)

```python
df = pd.read_csv('daily_electricity.csv', index_col='Date')

df
```

```python
# reset the index
df = df.reset_index()

df
```

```python
coal = alt.Chart(df).mark_line().encode(
    x='Date:T',
    y='Coal_MW:Q'
)

# Encoding Data Types
# https://altair-viz.github.io/user_guide/encoding.html#encoding-data-types

coal
```

```python
# Customize visualizations
# https://altair-viz.github.io/user_guide/customization.html

fig1 = alt.Chart(df).mark_line(
    color='red'
).encode(
    x='Date:T',
    y='Coal_MW:Q'
)

fig2 = alt.Chart(df).mark_line(
    color='blue'
).encode(
    x='Date:T',
    y='Gas_MW:Q'
)

alt_fig_all = fig1 + fig2

alt_fig_all.properties(
    title='Electricity Production'
)
```

# Transformations

```python
# Fold: convert wide-form data to long-form data
# https://altair-viz.github.io/user_guide/transform/fold.html?highlight=fold

cols = ['Coal_MW', 'Gas_MW', 'Hidroelectric_MW', 'Nuclear_MW', 'Wind_MW', 'Solar_MW', 'Biomass_MW']

#### FROM
# Date   Coal_MW   Gas_MW ...
# 2018   32.0      18.0

#### TO
# Date   key        value   
# 2018   Coal_MW    32.0    
# 2018   Gas_MW     18.0
# ...

multi_lc = alt.Chart(df).transform_fold(
    cols,
).mark_line().encode(
    x=alt.X('Date:T', title='Year'),
    y=alt.Y('value:Q', title='MW'),
    color='key:N'
).properties(
    title='Electricity Production',
    width=600,
    height=400
).interactive()

multi_lc
```
